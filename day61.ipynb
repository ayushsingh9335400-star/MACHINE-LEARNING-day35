```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
                                                
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import SGDRegressor

reg = SGDRegressor(
    penalty='l2',       # ridge-like regularization
    max_iter=500,
    eta0=0.1,
    learning_rate='constant',
    alpha=0.001,
    random_state=42
)
```

```python
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
print("R2 score", r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

# LASSO regression

```python
from sklearn.datasets import load_diabetes

import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt 
```

```python
data = load_diabetes()
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df['TARGET'] = data.target
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
```

```python
coefs = []

r2_scores = []

for i in[0,100,1000]:
    reg = Ridge(alpha=i)
    reg.fit(X_train,y_train)

    coefs.append(reg.coef_.tolist())
    y_pred = reg.predict(X_test)
    r2_scores.append(r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(14,9))
plt.subplot(221)
plt.bar(data.feature_names,coefs[0])
plt.title('alpha = 0, r2_score = {}'.format(round(r2_scores[0],2)))

plt.subplot(222)
plt.bar(data.feature_names,coefs[1])
plt.title('alpha = 0.1, r2_score = {}'.format(round(r2_scores[1],2)))

plt.subplot(223)
plt.bar(data.feature_names,coefs[2])
plt.title('alpha = 1, r2_score = {}'.format(round(r2_scores[2],2)))

plt.subplot(224)
plt.bar(data.feature_names,coefs[3])
plt.title('alpha = 10, r2_score = {}'.format(round(r2_scores[3],2)))

plt.show()
```

# Impact on Bias and Variance

```python
m = 100
X =  5 * np.random.rand(m, 1) - 2
y = 0.7 * X ** 2 - 2 * X + 3 + np.random.randn(m, 1)

plt.scatter(X,y)
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X.reshape(100, 1),y.reshape(100),test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=10)

X_train = poly.fit_transform(X_train)
X_test = poly.transform(X_test)
```

```python
from mlxtend.evalute import bias_variance_decomp

alphas = np.linespace(0,30,100)

loss = []
bias = []
variance = []
for i  in alphas:
    reg = Lasso(alpha=i)
    avg_expected_loss, avg_bias, avg_var = bias_varinace_decomp(
        reg, X_train, y_train, X_test, y_test,
        loss='mse',
        random_seed=123)
    loss.append(avg_expected_loss)
    bias.append(avg_bias)
    variance.append(avg_var)
        
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(alphas, loss, label="Loss")
plt.plot(alphas, bias, label="Bias^2")
plt.plot(alphas, variance, label="Variance")
plt.xlabel("Alpha")
plt.ylabel("Error")
plt.title("Bias-Variance Tradeoff (Lasso)")
plt.legend()
plt.show()
```

# Effect of Regularization on Loss Function 

```python
from sklearn.datasets import make_regression

X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=20,random_state=13)

plt.scatter(X,y)
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X,y)
print(reg.coef_)
print(reg.intercept_)
```

```python
def cal_loss(m,alpha):
    return np.sum((y - m*X.ravel() + 2.29)**2) + alpha*abs(m)
```

```python
def predict(m):
    return m*X - 2.29
```

```python
m = np.linspace(-45, 100, 100)
plt.figure(figsize=(12, 8))

for j in [0, 100, 500, 1000, 2500]:
    loss = []
    for i in range(m.shape[0]):
        loss_i = cal_loss(m[i], j)   # make sure this function is defined
        loss.append(loss_i)
    plt.plot(m, loss, label='alpha = {}'.format(j))  # outside the inner loop

plt.legend()
plt.xlabel('m (slope)')
plt.ylabel('Loss')
plt.title('Loss curves for different alpha values')
plt.show()
```
